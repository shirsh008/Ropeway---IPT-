```python
import geopandas as gpd
from shapely.geometry import Point
from scipy.spatial.distance import pdist, squareform
import numpy as np

# 1. Convert points and centroids to GeoDataFrames and project to EPSG:32644 (meters)
gdf_points = gpd.GeoDataFrame(
    df_3, geometry=gpd.points_from_xy(df_3.lon, df_3.lat), crs="EPSG:4326"
).to_crs("EPSG:32644")

gdf_centroids = gpd.GeoDataFrame(
    cluster_means_3, geometry=gpd.points_from_xy(cluster_means_3.lon, cluster_means_3.lat), crs="EPSG:4326"
).to_crs("EPSG:32644")

# 2. Calculate Mean Cluster Radius
gdf_points['centroid_geom'] = gdf_points['cluster'].map(gdf_centroids['geometry'])
gdf_points['distance_to_centroid'] = gdf_points.geometry.distance(gdf_points['centroid_geom'])
mean_radius = gdf_points['distance_to_centroid'].mean()
print(f"Mean Cluster Radius: {mean_radius:.2f} meters")

centroid_coords = np.array([(geom.x, geom.y) for geom in gdf_centroids.geometry])
dist_matrix = squareform(pdist(centroid_coords))

np.fill_diagonal(dist_matrix, np.inf)

min_dist = np.min(dist_matrix)
max_dist = np.max(dist_matrix[dist_matrix != np.inf])

print(f"Minimum Inter-Cluster Distance: {min_dist:.2f} meters")
print(f"Maximum Inter-Cluster Distance: {max_dist:.2f} meters")
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

X = df_3[['lat', 'lon']]

K_range = range(10, 100, 5)
inertia = []
silhouette = []

print("Calculating metrics for k-range. This may take a moment...")
for k in K_range:
    kmeans_model = KMeans(n_clusters=k, random_state=0, n_init='auto').fit(X)
    inertia.append(kmeans_model.inertia_)
    silhouette.append(silhouette_score(X, kmeans_model.labels_))

plt.style.use('seaborn-v0_8-whitegrid')
fig, ax1 = plt.subplots(1, 2, figsize=(14, 5))


ax1[0].plot(K_range, inertia, marker='o', color='#2c3e50', linewidth=2)
ax1[0].axvline(x=69, color='#e74c3c', linestyle='--', label='k=69 Selected')
ax1[0].set_title('Elbow Method for Optimal k', fontsize=14, fontweight='bold')
ax1[0].set_xlabel('Number of Clusters (k)', fontsize=12)
ax1[0].set_ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
ax1[0].legend()


ax1[1].plot(K_range, silhouette, marker='s', color='#27ae60', linewidth=2)
ax1[1].axvline(x=69, color='#e74c3c', linestyle='--', label='k=69 Selected')
ax1[1].set_title('Silhouette Score Evaluation', fontsize=14, fontweight='bold')
ax1[1].set_xlabel('Number of Clusters (k)', fontsize=12)
ax1[1].set_ylabel('Mean Silhouette Score', fontsize=12)
ax1[1].legend()

plt.tight_layout()
plt.savefig('k_selection_metrics.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
or_node = ox.distance.nearest_nodes(G, X=origin_log, Y=origin_lat)
des_node = ox.distance.nearest_nodes(G, X=work_log, Y=work_lat)

routes1 = []
missing_paths = 0

G_reversed = G.reverse(copy=False)

for d in des_node:
    try:

        paths_from_d = nx.single_source_dijkstra_path(G_reversed, source=d, weight='length')

        for o in or_node:
            if o in paths_from_d:
                correct_path = paths_from_d[o][::-1]
                routes1.append(correct_path)
            else:
                missing_paths += 1

    except nx.NetworkXNotImplemented:
        print("Graph issue encountered.")

route_lines2 = []

for path in routes1:
    if len(path) > 1:
        route_coords = [(G.nodes[n]['x'], G.nodes[n]['y']) for n in path]

        route_lines2.append(LineString(route_coords))
    else:
        print(f"Skipped a path because origin and destination are the same node: {path}")


gdf_routes = gpd.GeoDataFrame(geometry=route_lines2, crs="EPSG:4326")

orig_points = [Point(lon, lat) for lon, lat in zip(origin_log1, origin_lat1)]
dest_points = [Point(lon, lat) for lon, lat in zip(work_log, work_lat)]

gdf_orig = gpd.GeoDataFrame(geometry=orig_points, crs="EPSG:4326")
gdf_dest = gpd.GeoDataFrame(geometry=dest_points, crs="EPSG:4326")


print("Generating interactive map...")

m = gdf_routes.explore(
    color="blue",
    weight=3,
    opacity=0.6,
    name="Vehicle Routes"
)


m = gdf_dest.explore(
    m=m,
    color="green",
    marker_kwds={"radius": 6, "fill": True},
    name="Destinations"
)

folium.LayerControl().add_to(m)

m
```
